```python
# verification by self
import os
import pandas as pd

def filter_csv_files(path):
	ans = dict()
	# 获取指定路径下的所有文件和文件夹
	files = os.listdir(path)
	
	# 遍历所有文件
	for file in files:
		# 检查文件是否为csv文件
		if file.endswith(".csv"):
			# 构建文件的完整路径
			file_path = os.path.join(path, file)
			
			# 读取csv文件
			df = pd.read_csv(file_path)
			
			# 筛选出DATE = 2023-01-04的列
			filtered_df = df[df['DATE'] == '2023-01-04']
			ans[file[-7:-4]] = filtered_df
			
			# 打印筛选结果
			# print(f"File: {file}")
			# print(filtered_df)
			# print("--------------------")
	return ans

# 指定要遍历的路径
folder_path = "data/2023-data"

# 调用函数进行筛选
a = filter_csv_files(folder_path)
```
