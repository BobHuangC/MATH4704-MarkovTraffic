```python
def get_hour_k_step_transition_matrix(begin_hour, k):
	return (np.array([model.region_transition_matrix[i][j][begin_hour][k] for i in range(8) for j in range(8)]).reshape(8, 8))
```

```python
# atol = 1e-5
false_case_count = 0
count = 0
for begin_hour in range(0, 24):
    for k in range(1, 24):
        for l in range(1, 24):
            if begin_hour + k + l >= 24:
                continue
            count += 1
            
			# print(get_hour_k_step_transition_matrix(begin_hour, k))
            tmp_bool = np.allclose(
                get_hour_k_step_transition_matrix(begin_hour, k+l),
                np.dot(get_hour_k_step_transition_matrix(begin_hour, k), 
                       get_hour_k_step_transition_matrix(begin_hour + k, l)), 
                       )
            if not tmp_bool:
                print((begin_hour, k, l), ' this is the pair of (begin_hour, k, l)')
                false_case_count += 1

print(count, " this is the count")
print(false_case_count, " this is the false case count")
```
